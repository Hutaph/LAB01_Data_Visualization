## II. Tiền xử lý dữ liệu

### 1. Kiểm tra tính nhất quán dữ liệu

**Mục tiêu:** Loại bỏ các giá trị "rác" ẩn và đảm bảo các quy tắc logic cơ bản của dữ liệu.

Trước khi thực hiện các phân tích sâu hơn, chúng ta cần làm sạch các giá trị không hợp lệ phát sinh trong quá trình thu thập dữ liệu (crawl). Bước này đảm bảo rằng các phép toán thống kê và biểu đồ trên Dashboard sẽ phản ánh đúng bản chất của thị trường.

**Các bước thực hiện:**
- **Xử lý Placeholder:** Chuyển đổi các ký tự rác như `[]`, `{}`, `N/A` về giá trị `NaN` chuẩn của hệ thống để có thể tính toán chính xác tỷ lệ khuyết thiếu.
- **Kiểm tra Logic số học:** Đảm bảo các cột quan trọng như Giá (`price`), Điểm đánh giá (`rating`) và Lượt đánh giá (`reviews`) luôn ở định dạng số. Các giá trị không thể chuyển đổi sẽ được đưa về `NaN`.
- **Kiểm tra Logic Boolean:** Xác thực các biến logic (`is_prime`, `is_amazon_choice`,...) chỉ chứa các giá trị `True` hoặc `False`. Mọi giá trị khác lạ sẽ được coi là dữ liệu lỗi và đưa về `NaN`.

```python
placeholders = ["[]", "{}", "", "missing value", "N/A", "nan"]
df = df.replace(to_replace=placeholders, value=np.nan)
```

```python
print("--- Đang kiểm tra tính nhất quán dữ liệu ---")

for col in ['price', 'rating', 'reviews']:
    if col in df.columns:
        before_count = df[col].isnull().sum()

        df[col] = pd.to_numeric(df[col], errors='coerce')

        after_count = df[col].isnull().sum()
        
        print(f"Số lượng {col} không hợp lệ: {after_count - before_count}")
```

```python
target_bool_cols = ['is_amazon_choice', 'is_prime', 'has_variations', 'is_climate_friendly']
for col in target_bool_cols:
    if col in df.columns:
        inconsistent_count = df[~df[col].isin([True, False, np.nan])].shape[0]
        print(f"Số lượng {col} không hợp lệ (ngoài True/False): {inconsistent_count}")
        
        if inconsistent_count > 0:
            df.loc[~df[col].isin([True, False, np.nan]), col] = np.nan
```

### 2. Xử lý trùng lặp

Trong quá trình thu thập dữ liệu, một sản phẩm có thể xuất hiện nhiều lần do thuộc nhiều danh mục khác nhau. Việc tồn tại dữ liệu trùng lặp sẽ làm sai lệch các biểu đồ thống kê tổng quan trên Dashboard (như tổng doanh số, tổng số lượng sản phẩm).

**Các bước thực hiện:**
- **Kiểm tra trùng lặp theo mã ASIN:** Sử dụng mã định danh duy nhất (ASIN) để xác định các bản ghi lặp lại của cùng một mặt hàng.
- **Loại bỏ dữ liệu thừa:** Tiến hành xóa bỏ các dòng trùng lặp, chỉ giữ lại bản ghi đầu tiên xuất hiện để đảm bảo mỗi sản phẩm là duy nhất trong tập dữ liệu.

```python
duplicate_count = df.duplicated(subset=["asin"]).sum()
print(f"Số lượng bản ghi trùng lặp phát hiện (theo ASIN): {duplicate_count}")
```

```python
df = df.drop_duplicates(subset=["asin"], keep="first")
print(f"Đã loại bỏ trùng lặp. Kích thước dữ liệu hiện tại: {df.shape}")
```

### 3. Xử lý dữ liệu khuyết thiếu

Điền các giá trị mặc định cho dữ liệu trống để Dashboard hiển thị thân thiện hơn và loại bỏ các dòng thiếu thông tin cốt lõi để đảm bảo độ chính xác của biểu đồ.
- `original_price`: Điền bằng giá hiện tại (coi như không giảm giá).
- `sales_volume`, `delivery_info`: Điền các chuỗi mặc định ("0 bought...", "No delivery...").
- `price`, `rating`: Loại bỏ các dòng thiếu hai chỉ số quan trọng này.

```python
df["original_price"] = df["original_price"].fillna(df["price"])
df["sales_volume"] = df["sales_volume"].fillna("0 bought in past month")
df["delivery_info"] = df["delivery_info"].fillna("No delivery information")

df = df.dropna(subset=["price", "rating"])
print(f"Đã xử lý xong khuyết thiếu. Số lượng dòng còn lại: {len(df)}")
```

### 4. Kỹ thuật đặc trưng

**Mục tiêu:** Phân rã các thông tin văn bản thô thành các đặc trưng chi tiết phục vụ cho việc lọc và hiển thị trên Dashboard.
- `date`: Ngày dữ liệu được thu thập
- `discount`: Số tiền được giảm (Giá gốc - Giá hiện tại).
- `discount_rate`: Tỉ lệ phần trăm giảm giá.
- `sales_volume_num`: Con số sản lượng bán.
- `delivery_fee`: Chi phí vận chuyển (USD).
- `delivery_date_text`: Thời gian giao hàng dự kiến.
- `delivery_ship_to`: Quốc gia/vùng lãnh thổ nhận hàng.
- `delivery_stock_note`: Ghi chú về tình trạng kho hàng (ví dụ: "Only 5 left").

```python
df["date"] = "2026-03-24"
```

```python
df["discount"] = df["original_price"] - df["price"]
df["price"] = df["price"].clip(lower=0)

df["discount_rate"] = np.where(df["original_price"] > 0, df["discount"] / df["original_price"], 0.0)
```

```python
def extract_sales_num(value):
    s = str(value).lower()

    m = re.search(r"(\d+(?:\.\d+)?)\s*([kK]?)\+", s)
    if not m:
        return 0
    num = float(m.group(1))
    if m.group(2) == "k":
        num *= 1000
    return int(num)
```

```python
def extract_delivery_details(value):
    if pd.isna(value) or str(value).lower() == "no delivery information":
        return 0.0, pd.NA, pd.NA, pd.NA
    
    s = str(value).strip()

    fee = 0.0
    m_fee = re.search(r"\$(\d+(?:\.\d+)?)\s*delivery", s, flags=re.IGNORECASE)
    if m_fee: fee = float(m_fee.group(1))
    
    s_norm = re.sub(r"(\d)([A-Za-z])", r"\1 \2", s)
    
    m_date = re.search(r"delivery\s+(.*?)(?=\s*Ships to|$)", s_norm, flags=re.IGNORECASE)
    m_ship = re.search(r"Ships to\s+(.*?)(?=\s*Only\s+\d+\s+left|$)", s_norm, flags=re.IGNORECASE)
    m_stock = re.search(r"(Only\s+\d+\s+left.*)$", s_norm, flags=re.IGNORECASE)
    
    return (fee, 
            m_date.group(1).strip() if m_date else pd.NA, 
            m_ship.group(1).strip() if m_ship else pd.NA, 
            m_stock.group(1).strip() if m_stock else pd.NA)
```

```python
df["sales_volume_num"] = df["sales_volume"].apply(extract_sales_num)

delivery_res = df["delivery_info"].apply(extract_delivery_details)

df["delivery_fee"] = delivery_res.str[0]
df["delivery_date_text"] = delivery_res.str[1]
df["delivery_ship_to"] = delivery_res.str[2]
df["delivery_stock_note"] = delivery_res.str[3]
```
